```python
#
#                      SVM ALGORITHM
#
# The Linear Support Vector MAchine algorithm keeps in considerations two factors:
# 1) Hinge Loss function: similar to the perceptron alogirthm, this give a corrective value for theta when the prediction is not correct
#    In particular, we:
#     - compute  Z_i = y_i * ( (theta dot X_i )  + theta_zero)
#     - define the Loss as 0 if Z_i >= 1 
#     - define the loss as 1 - Z_i if the Z < 1
# 2) A normalization term given by the (L2 norm of theta) ^2
#    This term is related to the distance between the decision boundsand the margin bounds
#    By minimizing this term, we get larger margins
#
# The objective function resulting is given by:
#  - the average of of the Hinge Loss (i.e. sum of all the Loss divided by N)
#  - lambda/2 * the normalization term
img = Image.open('pictures/SVM objective function.png')
display(img)
```
