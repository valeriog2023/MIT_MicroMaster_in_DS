```python
#
#  RANDOMIZATION
# 
def randomization(n: int)-> np.array:
    """
    Arg:
      n - an integer
    Returns:
      A - a randomly-generated nx1 Numpy array.
    """

    return np.random.random([n,1])

assert isinstance(randomization(3),type(np.array([[1],[2],[3]])))
assert randomization(3).shape == (3,1)
```

```python
# Here, we will write a function neural_network, which will apply a neural network operation with 2 inputs 
# and 1 output and a given weight matrix.
#

def neural_network(inputs:np.array(), weights:np.array()) -> np.array:
    """
     Takes an input vector and runs it through a 1-layer neural network
     with a given weight matrix and returns the output.

     Arg:
       inputs - 2 x 1 NumPy array
       weights - 2 x 1 NumPy array
     Returns (in this order):
       out - a 1 x 1 NumPy array, representing the output of the neural network (use tanh function)
    """
    #Your code here
    assert inputs.shape == (2,1)
    assert weights.shape == (2,1)
    return np.tanh(weights.T.dot(inputs))

x=np.array([[100],[200]])
w=np.array([[1],[2]])
z = neural_network(x,w)

assert z.shape == (1,1)


```

```python
#
# You can use numpy to run a function in parallel over all the elements of one or more array:
# 
#   1) define you function to take single values/scalars as input
#   2) convert it to a vector function with: np.vectorize 
#
def scalar_function(x:float, y:float) -> float:
    """
    Returns the f(x,y) defined as 
       x * y if x =< y
       x/y     otherwise

    where x and y are scalars      
    """
    if x > y:
        return  x/y
    else:
        return  x * y
    

def vector_function(x:np.array, y:np.array) -> np.array:
    """
    Make sure vector_function can deal with vector input x,y 
    """
    v_f = np.vectorize(scalar_function)
    return v_f(x,y)

assert all(vector_function(np.array([1,2,3,10]),  np.array([5,5,5,5])) == np.array([ 5, 10, 15,  2]))
    
```
